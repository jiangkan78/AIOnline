## 01. 了解机器视觉环境

### 1.1 机器视觉

   机器视觉也就是机器人的视觉系统，它与人的眼睛有所不同，但是有相似之处。人眼能看到物体是因为物体反射的光线刺激人眼的感觉细胞，通过视觉神经在大脑中形成成像。而机器人的视觉是通过摄像头的光敏元件将信号转化成数字信号并将其量化为数字矩阵，表示物体的反射光的强弱。
   
   机器视觉通过计算机编程，设计算法来分析图像中有什么，把分析的信息用于某些特殊的任务当中，因此我们把机器视觉系统看作是一个与周围环境进行交互的大的实体的一部分，它反馈环境中的一些有用信息给其他部分，其他部分使用这些信息做出决策和执行这些决策，如在图像搜索中根据条件搜索出关键信息反馈给系统，在机器人导航中视觉反馈路线信息给系统指导机器人行进等其他操作，医学图像分析领域可以协助医生分析照片中反应的病理信息等等。
   
   机器视觉的特点就是提高了生产的柔性和自动化程度，在一些不适合人工作业的危险工作环境代替人工的视觉；同时可以应用在工业生产过程中，检查产品质量等，而且机器视觉易于实现信息集成，也是实现计算机集成制造的关键基础技术。
   
### 1.2 像素
   讲到图像就不得不先说说像素，什么是像素呢？首先要说的是每个图像都是由像素组成的，像素是图像的原始构建块，没有比像素更精细的粒度了。
   
   通常情况下，我们认为像素是出现在图像中的某一个特定位置的光的“颜色”或“强度”。
   
   如果我们把一个图像看做是一个网格，网络中的每个方块就是一个独立的像素。比如我们有一个张500x300的图片，这就意味着这个像素的网格是500行和300列的，总共有500×300=150,000像素在这个图片里。大多数像素使用两种方式来表示：灰度和颜色。
   
   在灰度图像中，每个像素都是介于0和255之间，0代表黑色，255代表白色。其中越接近0就越深，越接近255就越浅。
   
   颜色像素通常使用RGB表示。R表示红色，G表示绿色，B表示蓝色，也存在其它颜色空间。这三个颜色值都是0和255之间。白色为(255,255,255)，黑色为(0,0,0)，纯色红色就是(255,0,0)，等等。

## 02 第一个程序显示图片

通过调用opencv的方法显示本地图片。

```python
import cv2

image = cv2.imread("..\pic\pic1.png")
cv2.imshow("Image window", image)

key = cv2.waitKey()

if key != 1 :
    print("触发了按键！")

cv2.destroyAllWindows()
```

#### 关于程序方法说明
以上程序上使用了4个关键方法：

* **imread**，读取图像OpenCV提供了方法cv2.imread来读取图像，该方法支持各种静态图像格式。
    * 该方法的语法格式为：retval=cv2.imread(filename[,flags])
      * retval 是返回值，其值就是读取到的图像，如果没有读取到图像，则返回“None”。
      * filename表示要读取的图像名字。
      * flags是读取标志。没有特殊要求可以不赋值，默认读取出的格式是BGR格式。
* **imshow**，为显示图像使用的方法
    * 该方法的语法格式为：None = cv2.imshow(winname,mat)
      * winname为窗口名称
      * mat为要显示的图像
* **waitKey**，等待按键使用cv2.waitKey()方法，当用户按下键盘后，该语句会被执行，并获取返回值。
    * 该方法的语法格式为：retval = cv2.waitKey([delay])
      * retval 是返回值。如果没有按键被按下，则返回-1，如果有按键被按下，则返回按键的ASCII码。
      * delay表示等待的时间，单位为毫秒，如果为-1或0比欧式无限等待，默认为0。
* **destroyAllWindows**，释放资源

后续为了方便演示我们会使用matlib来显示所有图像，后面会使用到如下方法：

* **cvtColor**，完成颜色空间的转换后续的课程会详细讲解
* **imshow**，这个是matlib的显示图像方法与opencv的imshow功能相同


```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread("..\pic\pic1.png")
equal = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)#opencv 返回的图像是BGR但是matlib显示图像是RGB所以需要进行转换
fig = plt.figure(figsize=(20,10), dpi=100)
plt.imshow(equal)
```

## 04 像素的操作

经过前面例子，我们已经学会图片的加载和保存，那么我们怎么访问实际的像素数据呢？
  1. 我们首先要记住的一点就是OpenCV所代表的图像就是一个网格，一个矩阵，它可以用NumPy数组来表示。NumPy（Numerical Python）是 Python 中的一个线性代数库。对数组执行数学运算和逻辑运算时，NumPy 是非常有用的。在用 Python 对 n 维数组和矩阵进行运算时，NumPy 提供了大量有用的操作方法。尤其是Numpy.array库，Numpy.array库是Python处理图像的基础。
  2. 图像通常分为二值图，灰度图和彩色图，下面我们来依次介绍一下这三种图以及表示方法。这些内容有助于我们理解像素的操作。

### 4.1 二值图像
二值图像是指仅包含黑色和白色两种颜色的图像。在计算机中，通过一个矩阵来表示和处理图像。例如，图中所示的一个字母A的图像，计算机在处理该图像时，会首先将其划分为一个个的小方块，每一个小方块就是一个独立的处理单位，称为像素点。接下来，计算机会将其中的白色像素点（白色小方块区域）处理为“1”，将黑色像素点（黑色小方块区域）处理为“0”，以方便进行后续的存储和处理等操作。

### 4.3 彩色图像
相比二值图像和灰度图像，彩色图像是更常见的一类图像，它能表现更丰富的细节信息。从不同角度采用不同的方式表述颜色的模式称为色彩空间，或者颜色空间、颜色模式等，我们今天只介绍常用的RGB颜色空间。

在 RGB 颜色空间中，存在 R（red，红色）通道、G（green，绿色）通道和 B（blue，蓝色）通道，共三个通道。每个色彩通道值的范围都在[0,255]之间，我们用这三个色彩通道的组合表示颜色。就如同我们有三个油漆桶，分别装了红色、绿色、蓝色的油漆，我们分别从每个油漆桶中取容量为0~255个单位的不等量的油漆，将三种油漆混合就可以调配出一种新的颜色。三种油漆经过不同的组合，共可以调配出所有常见的256×256×256=16 777 216种颜色。

## 05 绘图

### 5.1 画线
画线OpenCV提供了方法**cv2.line()**用来绘制直线，该方法的语法格式为：img = cv2.line(img,pt1,pt2,color[,thickness[,lineType]])

* img 其值就是要绘制的图像载体，也可以成为画板。
* pt1表示线段的第一个点，起点。
* pt2表示线段的第二个点，终点。
* color表示绘制形状的颜色。通常使用BGR模型表示。
* thickness表示线宽
* lineType 表示线条类型。通常使用默认即可。

```python
image = cv2.imread("..\pic\pic4.png")
cv2.line(image,(0,0),(680,480),(0,0,255),3)
equal = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

颜色为RGB格式所以0,0,255为红色。就如前面提到的OpenCV实际存储时是BGR，也就是说如果要显示红色为0,0，255，蓝色为255,0,0。

### 5.2 画矩形
OpenCV提供了函数**cv2.rectangle（）**用来绘制矩形。该函数的语法格式为：img=cv2.rectangle（img,pt1,pt2,color[,thickness[,lineType]]）
* pt1为矩形顶点
* pt2为矩形中与pt1对角的顶点。
* color为颜色
* thickness为线条的粗细。默认值是1，如果设置为-1，表示填充图形（即绘制的图形是实心的）。
* lineType：线条的类型,可以不用考虑。

```python
cv2.rectangle(image,(0,0),(30,30),(0,123,0),-1)
equal = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

### 5.3 画圆形
OpenCV提供了函数**cv2.circle（）**用来绘制圆。该函数的语法格式为：img=cv2.circle（img,center,radius,color[,thickness[,lineType]]）
* 参数img、color、thickness、lineType的含义如前面的说明所示。
* center为圆心。
* radius为半径。

```python
cv2.circle(equal,(50,50),40,(0,0,255),-1)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

### 5.4 写文字
OpenCV提供了函数**cv2.putText（）**用来在图形上绘制文字。该函数的语法格式为：
img=cv2.putText（img,text,org,fontFace,fontScale,color[,thickness[,lineType[,bottomLeftOrigin]]]）

* 参数img、color、thickness、lineType和shift的含义如前面的说明所示。
* text为要绘制的字体。
* org为绘制字体的位置，以文字的左下角为起点。
* fontFace表示字体类型，其参数类型及含义如表19-2所示。
* fontScale表示字体大小。
* bottomLeftOrigin用于控制文字的方向。默认值为False，当设置为True时，文字是垂直镜像的效果。

```python
image = cv2.imread("..\pic\pic4.png")
equal = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.putText(equal,'Hello OpenCV',(0,70),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,0.8,(255,0,0),2)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

 OpenCV提供了方法**capture.read()**读取摄像头数据。该方法的语法格式为 img = capture.read()
 * img为返回的图像数据

### 6.3 释放资源
OpenCV提供了方法**capture.release()**释放摄像头。该方法的语法格式为：None = capture.release()


```python
import cv2
import matplotlib.pyplot as plt

capture = cv2.VideoCapture(0)
while(True):
    ret, frame = capture.read()
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):
        break;
capture.release()
cv2.destroyAllWindows()
```

## 07 Numpy库的使用
NumPy(Numerical Python) 是 Python 语言的一个扩展程序库，支持大量的维度数组与矩阵运算，此外也针对数组运算提供大量的数学函数库。
NumPy 是一个运行速度非常快的数学库，主要用于数组计算，包含：

 1. 一个强大的N维数组对象 ndarray
 2. 广播功能函数
 3. 整合 C/C++/Fortran 代码的工具
 4. 线性代数、傅里叶变换、随机数生成等功能
 
我们在后面的程序中要使用到np.random.randint，接口原型是numpy.random.randint(low, high=None, size=None, dtype='l')。
* low: int类型生成的数值最低要大于等于low;
* high: int类型 (可选)如果使用这个值，则生成的数值在\[low, high\]区间；
* size: int or tuple of ints类型(可选)，输出随机数的尺寸；
* dtype(可选) 想要输出的格式。如int64、int等等。

### 7.1 绘制随机圆形

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("..\pic\pic4.png")
for i in range(0, 25):
    radius = np.random.randint(5, high=60)
    color = np.random.randint(0, high=256, size=(3,)).tolist()
    pt = np.random.randint(0, high=150, size=(2,))
    cv2.circle(image, (pt[0],pt[1]), radius, color, -1)
    
equal = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

### 7.2 绘制随机矩形

## 08 任务画国旗

* 德国国旗思路：
  * 我们需要先定义一个画板，可以使用numpy来完成。比如300*300的画板里画国旗。德国国旗是三个矩形。所以只要定义好起点和终点就可以了。
    1. 指定第一个黑色矩形的起点和终点比如宽定义为150，高为34.任意选择一个坐标点为起点（50,100）就可以根据宽和高计算出终点坐标为（200,134）
    2. 第一个矩形的左下角为第二个红色矩形的起点（50,134），终点坐标根据宽高就能计算得出（200,168）。以此类推出第三个黄色矩形的起点为（50,168）终点为（200,202）
    3. 最后在显示一个Germany的文字在国旗下方。

```python
import cv2
import numpy as np

canvas = np.zeros((300,300,3),dtype="uint8")
canvas[0:300,0:300] = (255,255,255)

cv2.rectangle(canvas,(50,100),(200,134),(0,0,0),-1)
cv2.rectangle(canvas,(50,134),(200,168),(0,0,255),-1)
cv2.rectangle(canvas,(50,168),(200,202),(0,255,255),-1)
cv2.putText(canvas,'Germany',(60,230),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)

equal = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```

  * 我们还是需要先定义一个画板，可以使用numpy来完成。比如300*300的画板里画国旗。日本国旗是一个矩形中间有个圆形。所以还是要先定义好大矩形的起点和终点，然后在矩形的中间位置画个圆就可以了
    1. 指定第一个黑色矩形的起点和终点比如宽定义为200，高为100.任意选择一个坐标点为起点（50,100）就可以根据宽和高计算出终点坐标为（250,200）
    2. 根据矩形的长宽坐标可以计算出矩形的中心为（150,150）即为圆形的圆心坐标，再定义半径为30，就可以画出日本国旗了。
    3. 最后在显示一个Japan的文字在国旗下方。

```python
'''
在300*300的画布中间画一个width 200 height 100的日本国旗
'''
import cv2
import numpy as np


canvas = np.zeros((300,300,3),dtype="uint8")

cv2.rectangle(canvas,(50,100),(250,200),(255,255,255),-1)
cv2.circle(canvas,(150,150),30,(0,0,255),-1)
cv2.putText(canvas,'Japan',(100,230),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
equal = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(20,10), dpi=100)
plt.subplot(221)
plt.imshow(equal)
```
